```
#MRO (C3 Linearization Algorithm)
class F: pass
class A(F): pass
class B(F): pass
class C(A, B): pass
class D(A, B): pass
class E(C, D): pass

print(E.__mro__)
```

```
#Generators
'''These are kinda functions that store their state.
Usecase is similar to iterators but much simpler to build.'''

def gen_fib():
  a, b = 0, 1
  while True:
    yield a
    a, b = b, a + b
    
x = gen_fib()
print(next(x))
print(next(x))
```

```
#Data structures in Python
from collections import deque, OrderedDict, Counter

q = deque()
q.append(4)
q.append(3)
print(q.pop())
print(q.pop())

c = Counter("oskghdnzk.jms,ndgskf")
print(c.most_common(2))

o = OrderedDict()
o['a'] = 'b'
o['z'] = 'x'
o['b'] = 'q'

print(o)
del o['z']
print(o)
```
